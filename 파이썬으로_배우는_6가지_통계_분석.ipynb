```
# íŒŒì´ì¬ìœ¼ë¡œ ì§ì ‘ ì‹¤í–‰í•´ë³´ëŠ” 6ê°€ì§€ í†µê³„ ë¶„ì„ ì‹œë‚˜ë¦¬ì˜¤
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
import scipy.stats as stats
import numpy as np
import pandas as pd

print("ì•ˆë…•í•˜ì„¸ìš”! 'ë‚˜ì˜ í†µê³„ ì„ ìƒë‹˜'ê³¼ í•¨ê»˜í•˜ëŠ” íŒŒì´ì¬ í†µê³„ ë¶„ì„ ì‹œê°„ì…ë‹ˆë‹¤. ğŸš€\n")

# --- ì‹œë‚˜ë¦¬ì˜¤ 1: ì¼í‘œë³¸ T-ê²€ì • (One-Sample T-test) ---
print("="*50)
print("â˜• ì‹œë‚˜ë¦¬ì˜¤ 1: ìš°ë¦¬ ì¹´í˜ ì†ë‹˜ë“¤ì€ ì •ë§ 10,000ì› ì´ìƒ ì“¸ê¹Œ?")
print("="*50)

# ë°ì´í„°: í•œ ê·¸ë£¹ì˜ êµ¬ë§¤ ê¸ˆì•¡
purchase_amounts = [12000, 12500, 13000, 12800, 12600, 12200, 12700, 12900, 12400, 12100]
# ë¹„êµí•˜ë ¤ëŠ” ëª©í‘œ í‰ê·  êµ¬ë§¤ ê¸ˆì•¡
expected_value = 10000

# ì¼í‘œë³¸ T-ê²€ì • ìˆ˜í–‰
# ttest_1samp í•¨ìˆ˜ëŠ” ë°ì´í„°ì™€ ë¹„êµí•  í‰ê· ê°’ì„ ë°›ì•„ì„œ T-í†µê³„ëŸ‰ê³¼ p-valueë¥¼ ëŒë ¤ì¤ë‹ˆë‹¤.
t_stat_1, p_value_1 = stats.ttest_1samp(purchase_amounts, expected_value)

# ê²°ê³¼ ì¶œë ¥
print(f"T-í†µê³„ëŸ‰ (T-statistic): {t_stat_1:.4f}")
print(f"p-value: {p_value_1:.10f}")

# í•´ì„
alpha = 0.05  # ìœ ì˜ìˆ˜ì¤€ 5%
if p_value_1 < alpha:
    print(f"ê²°ë¡ : ê·¸ë£¹ì˜ í‰ê·  êµ¬ë§¤ ê¸ˆì•¡ì´ {expected_value}ì›ê³¼ ìœ ì˜ë¯¸í•˜ê²Œ ë‹¤ë¦…ë‹ˆë‹¤. (ìš°ì—°ì´ ì•„ë‹ˆì•¼!)")
else:
    print(f"ê²°ë¡ : ê·¸ë£¹ì˜ í‰ê·  êµ¬ë§¤ ê¸ˆì•¡ì´ {expected_value}ì›ê³¼ ìœ ì˜ë¯¸í•˜ê²Œ ë‹¤ë¥´ì§€ ì•ŠìŠµë‹ˆë‹¤. (ìš°ì—°ì¼ ìˆ˜ ìˆì–´!)")
print("\n")


# --- ì‹œë‚˜ë¦¬ì˜¤ 2: ëŒ€ì‘í‘œë³¸ T-ê²€ì • (Paired Sample T-test) ---
print("="*50)
print("ğŸ›ï¸ ì‹œë‚˜ë¦¬ì˜¤ 2: í”„ë¡œëª¨ì…˜, ì •ë§ íš¨ê³¼ê°€ ìˆì—ˆì„ê¹Œ?")
print("="*50)

# ë°ì´í„°: ë˜‘ê°™ì€ ê³ ê°ë“¤ì˜ í”„ë¡œëª¨ì…˜ ì „í›„ êµ¬ë§¤ ê¸ˆì•¡
before_promo = [12000, 12500, 13000, 12800, 12600, 12200, 12700, 12900, 12400, 12100]
after_promo = [13500, 14000, 13800, 13600, 13900, 13700, 13400, 13200, 14100, 13300]

# ëŒ€ì‘í‘œë³¸ T-ê²€ì • ìˆ˜í–‰
# ttest_rel í•¨ìˆ˜ëŠ” 'ì „' ë°ì´í„°ì™€ 'í›„' ë°ì´í„°ë¥¼ ë°›ì•„ì„œ T-í†µê³„ëŸ‰ê³¼ p-valueë¥¼ ëŒë ¤ì¤ë‹ˆë‹¤.
t_stat_2, p_value_2 = stats.ttest_rel(before_promo, after_promo)

# ê²°ê³¼ ì¶œë ¥
print(f"T-í†µê³„ëŸ‰ (T-statistic): {t_stat_2:.4f}")
print(f"P-value: {p_value_2:.10f}")

# í•´ì„
if p_value_2 < alpha:
    print("ê²°ë¡ : í”„ë¡œëª¨ì…˜ ì „í›„ êµ¬ë§¤ ê¸ˆì•¡ ì°¨ì´ê°€ ìœ ì˜ë¯¸í•©ë‹ˆë‹¤. (í”„ë¡œëª¨ì…˜ íš¨ê³¼ê°€ ìˆì—ˆì–´!)")
else:
    print("ê²°ë¡ : í”„ë¡œëª¨ì…˜ ì „í›„ êµ¬ë§¤ ê¸ˆì•¡ ì°¨ì´ê°€ ìœ ì˜ë¯¸í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. (íš¨ê³¼ê°€ ìˆë‹¤ê³  ë³´ê¸° ì–´ë ¤ì›Œ!)")
print("\n")


# --- ì‹œë‚˜ë¦¬ì˜¤ 3: ë…ë¦½í‘œë³¸ T-ê²€ì • (Independent Samples T-test) ---
print("="*50)
print("ğŸ†š ì‹œë‚˜ë¦¬ì˜¤ 3: A í”„ë¡œëª¨ì…˜ vs B í”„ë¡œëª¨ì…˜, ëˆ„ê°€ ë” ë‚˜ì„ê¹Œ?")
print("="*50)

# ë°ì´í„°: ì„œë¡œ ë‹¤ë¥¸ ë‘ ê·¸ë£¹ì˜ êµ¬ë§¤ ê¸ˆì•¡
promotion_A = [12000, 12500, 13000, 12800, 12600, 12200, 12700, 12900, 12400, 12100]
promotion_B = [13500, 14000, 13800, 13600, 13900, 13700, 13400, 13200, 14100, 13300]

# ë…ë¦½í‘œë³¸ T-ê²€ì • ìˆ˜í–‰ (ë“±ë¶„ì‚°ì„ ê°€ì •í•˜ì§€ ì•ŠìŒ)
# ttest_ind í•¨ìˆ˜ëŠ” ë‘ ê·¸ë£¹ì˜ ë°ì´í„°ë¥¼ ë°›ìŠµë‹ˆë‹¤. equal_var=FalseëŠ” ë‘ ê·¸ë£¹ì˜ ë°ì´í„° í©ì–´ì§ ì •ë„ê°€ ë‹¬ë¼ë„ ê´œì°®ë‹¤ê³  ì•Œë ¤ì£¼ëŠ” ì˜µì…˜ì…ë‹ˆë‹¤.
t_stat_3, p_value_3 = stats.ttest_ind(promotion_A, promotion_B, equal_var=False)

# ê²°ê³¼ ì¶œë ¥
print(f"T-í†µê³„ëŸ‰ (T-statistic): {t_stat_3:.4f}")
print(f"P-value: {p_value_3:.10f}")

# í•´ì„
if p_value_3 < alpha:
    print("ê²°ë¡ : A ê·¸ë£¹ê³¼ B ê·¸ë£¹ì˜ êµ¬ë§¤ ê¸ˆì•¡ ì°¨ì´ê°€ ìœ ì˜ë¯¸í•©ë‹ˆë‹¤. (ë‘ í”„ë¡œëª¨ì…˜ì˜ íš¨ê³¼ê°€ ë‹¬ë¼!)")
else:
    print("ê²°ë¡ : A ê·¸ë£¹ê³¼ B ê·¸ë£¹ì˜ êµ¬ë§¤ ê¸ˆì•¡ ì°¨ì´ê°€ ìœ ì˜ë¯¸í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. (íš¨ê³¼ ì°¨ì´ê°€ ìˆë‹¤ê³  ë§í•˜ê¸° í˜ë“¤ì–´!)")
print("\n")


# --- ì‹œë‚˜ë¦¬ì˜¤ 4: ë¶„ì‚° ë¶„ì„ (ANOVA) ---
print("="*50)
print("ğŸ“Š ì‹œë‚˜ë¦¬ì˜¤ 4: ë¬´ë£Œë°°ì†¡, 10%í• ì¸, 1+1 í–‰ì‚¬ ì¤‘ ìµœê³ ëŠ”?")
print("="*50)

# ë°ì´í„°: ì„¸ ê°€ì§€ í”„ë¡œëª¨ì…˜ë³„ êµ¬ë§¤ ê¸ˆì•¡
free_shipping = [12000, 12500, 13000, 12800, 12600]
discount_10 = [13500, 14000, 13800, 13600, 13900]
buy1get1 = [15000, 15500, 15200, 14800, 15400]

# ANOVA ìˆ˜í–‰
# f_oneway í•¨ìˆ˜ëŠ” ì„¸ ê°œ ì´ìƒì˜ ê·¸ë£¹ ë°ì´í„°ë¥¼ ë°›ì•„ì„œ F-í†µê³„ëŸ‰ê³¼ p-valueë¥¼ ëŒë ¤ì¤ë‹ˆë‹¤.
f_stat_4, p_value_4 = stats.f_oneway(free_shipping, discount_10, buy1get1)

# ê²°ê³¼ ì¶œë ¥
print(f"F-í†µê³„ëŸ‰ (F-statistic): {f_stat_4:.4f}")
print(f"P-value: {p_value_4:.10f}")

# í•´ì„
if p_value_4 < alpha:
    print("ê²°ë¡ : í”„ë¡œëª¨ì…˜ ê·¸ë£¹ ê°„ êµ¬ë§¤ ê¸ˆì•¡ ì°¨ì´ê°€ ìœ ì˜ë¯¸í•©ë‹ˆë‹¤. (ì ì–´ë„ í•˜ë‚˜ëŠ” íš¨ê³¼ê°€ ë‹¬ë¼!)")
else:
    print("ê²°ë¡ : í”„ë¡œëª¨ì…˜ ê·¸ë£¹ ê°„ êµ¬ë§¤ ê¸ˆì•¡ ì°¨ì´ê°€ ì—†ìŠµë‹ˆë‹¤. (ë‹¤ë“¤ ë¹„ìŠ·ë¹„ìŠ·í•´!)")
print("\n")


# --- ì‹œë‚˜ë¦¬ì˜¤ 5: ì¹´ì´ì œê³± ê²€ì • (Chi-Square Test) ---
print("="*50)
print("ğŸ“ˆ ì‹œë‚˜ë¦¬ì˜¤ 5: A/B í”„ë¡œëª¨ì…˜, ì–´ëŠ ìª½ ì „í™˜ìœ¨ì´ ë” ë†’ì„ê¹Œ?")
print("="*50)

# ë°ì´í„°: A/B í”„ë¡œëª¨ì…˜ë³„ [ì „í™˜í•œ ê³ ê° ìˆ˜, ì „í™˜í•˜ì§€ ì•Šì€ ê³ ê° ìˆ˜]
promotion_A_counts = [60, 440]
promotion_B_counts = [90, 410]

# ì¹´ì´ì œê³± ê²€ì •ì„ ìœ„í•œ ë¶„í• í‘œ(contingency table) ìƒì„±
conversion_table = np.array([promotion_A_counts, promotion_B_counts])

# ì¹´ì´ì œê³± ê²€ì • ìˆ˜í–‰
# chi2_contingency í•¨ìˆ˜ëŠ” ë¶„í• í‘œë¥¼ ë°›ì•„ì„œ ì¹´ì´ì œê³± í†µê³„ëŸ‰, p-value, ììœ ë„(dof), ê¸°ëŒ€ë¹ˆë„í‘œë¥¼ ëŒë ¤ì¤ë‹ˆë‹¤.
chi2_stat_5, p_value_5, dof_5, expected_5 = stats.chi2_contingency(conversion_table)

# ê²°ê³¼ ì¶œë ¥
print(f"ì¹´ì´ì œê³± í†µê³„ëŸ‰ (Chi-Square Statistic): {chi2_stat_5:.4f}")
print(f"P-value: {p_value_5:.10f}")
print(f"ììœ ë„ (dof): {dof_5}")

# í•´ì„
if p_value_5 < alpha:
    print("ê²°ë¡ : A/B í”„ë¡œëª¨ì…˜ ê°„ ì „í™˜ìœ¨ ì°¨ì´ê°€ ìœ ì˜ë¯¸í•©ë‹ˆë‹¤. (ë‘ í”„ë¡œëª¨ì…˜ì˜ ì„±ê³¼ê°€ ë‹¬ë¼!)")
else:
    print("ê²°ë¡ : A/B í”„ë¡œëª¨ì…˜ ê°„ ì „í™˜ìœ¨ ì°¨ì´ê°€ ìœ ì˜ë¯¸í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. (ì„±ê³¼ ì°¨ì´ê°€ ìš°ì—°ì¼ ìˆ˜ ìˆì–´!)")
print("\n")


# --- ì‹œë‚˜ë¦¬ì˜¤ 6: ì¿ í‚¤ìº£ A/B í…ŒìŠ¤íŠ¸ (ë¹„ëª¨ìˆ˜ ê²€ì • ë° ë¦¬í…ì…˜ ë¶„ì„) ---
print("="*50)
print("ğŸ® ì‹œë‚˜ë¦¬ì˜¤ 6: ì¿ í‚¤ìº£ ê²Œì„, ì²« ê´€ë¬¸ì€ ì–´ë””ê°€ ì¢‹ì„ê¹Œ?")
print("="*50)

# ì‹¤ì œ ë°ì´í„° íŒŒì¼ ëŒ€ì‹ , ë¹„ìŠ·í•œ ëª¨ì–‘ì˜ ê°€ìƒ ë°ì´í„°(DataFrame)ë¥¼ ë§Œë“­ë‹ˆë‹¤.
# ë°ì´í„°ë¥¼ '/content/cookie_cats.csv' íŒŒì¼ì—ì„œ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
df = pd.read_csv('/content/cookie_cats.csv')

# 1. ê²Œì„ ë¼ìš´ë“œ ìˆ˜ ë¹„êµ (Mann-Whitney U ê²€ì •)
# versionì— ë”°ë¼ ë‘ ê·¸ë£¹ìœ¼ë¡œ ë°ì´í„°ë¥¼ ë‚˜ëˆ•ë‹ˆë‹¤.
gate_30 = df[df["version"] == "gate_30"]["sum_gamerounds"]
gate_40 = df[df["version"] == "gate_40"]["sum_gamerounds"]

# Mann-Whitney U ê²€ì • ìˆ˜í–‰
# mannwhitneyu í•¨ìˆ˜ëŠ” ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ì§€ ì•Šì„ ê²ƒ ê°™ì€ ë‘ ë…ë¦½ ê·¸ë£¹ì„ ë¹„êµí•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
stat_6, p_value_6_mw = stats.mannwhitneyu(gate_30, gate_40, alternative='two-sided')

print("--- ê²Œì„ ë¼ìš´ë“œ ìˆ˜ ë¹„êµ (Mann-Whitney U ê²€ì •) ---")
print(f"p-value: {p_value_6_mw:.4f}")
if p_value_6_mw < alpha:
    print("ê²°ë¡ : ë‘ ê·¸ë£¹ ê°„ ê²Œì„ ë¼ìš´ë“œ ìˆ˜ì— í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•œ ì°¨ì´ê°€ ìˆìŠµë‹ˆë‹¤.\n")
else:
    print("ê²°ë¡ : ë‘ ê·¸ë£¹ ê°„ ê²Œì„ ë¼ìš´ë“œ ìˆ˜ì— í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•œ ì°¨ì´ê°€ ì—†ìŠµë‹ˆë‹¤.\n")


# 2. 1ì¼ì°¨ ë¦¬í…ì…˜(ì¬ë°©ë¬¸ìœ¨) ë¹„êµ (ì¹´ì´ì œê³± ê²€ì •)
# pd.crosstabì€ ë²„ì „ë³„, ë¦¬í…ì…˜ë³„ë¡œ ì‚¬ëŒ ìˆ˜ë¥¼ ì„¸ì–´ í‘œë¡œ ë§Œë“¤ì–´ì¤ë‹ˆë‹¤.
retention_table = pd.crosstab(df['version'], df['retention_1'])
print("--- 1ì¼ì°¨ ë¦¬í…ì…˜ ë¶„í• í‘œ ---")
print(retention_table)
print("")

# ì¹´ì´ì œê³± ê²€ì • ìˆ˜í–‰
chi2_6, p_value_6_chi, _, _ = stats.chi2_contingency(retention_table)

print("--- 1ì¼ì°¨ ë¦¬í…ì…˜ ë¹„êµ (ì¹´ì´ì œê³± ê²€ì •) ---")
print(f"p-value: {p_value_6_chi:.4f}")
if p_value_6_chi < alpha:
    print("ê²°ë¡ : ë‘ ê·¸ë£¹ ê°„ 1ì¼ì°¨ ë¦¬í…ì…˜ì— ìœ ì˜ë¯¸í•œ ì°¨ì´ê°€ ìˆìŠµë‹ˆë‹¤.")
else:
    print("ê²°ë¡ : ë‘ ê·¸ë£¹ ê°„ 1ì¼ì°¨ ë¦¬í…ì…˜ì— ìœ ì˜ë¯¸í•œ ì°¨ì´ê°€ ì—†ìŠµë‹ˆë‹¤.")
```
