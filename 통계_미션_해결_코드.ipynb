```
# ğŸ› ï¸ ì´ˆê¸° ì„¤ì •
import pandas as pd
import numpy as np

import seaborn as sns
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score, classification_report, confusion_matrix
from scipy import stats

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# í•œê¸€ í°íŠ¸ ì„¤ì •
# Colab, Jupyter Notebook ë“± í™˜ê²½ì— ë§ê²Œ í°íŠ¸ ì„¤ì •ì„ í•´ì£¼ì„¸ìš”.
# ì˜ˆì‹œ: ë‚˜ëˆ”ê³ ë”•
try:
    fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
    if fm.fontManager.findfont(fontpath):
        plt.rc('font', family='NanumBarunGothic')
        print("ë‚˜ëˆ”ë°”ë¥¸ê³ ë”• í°íŠ¸ ì„¤ì • ì‹œë„ ì™„ë£Œ.")
    else:
        print("ë‚˜ëˆ”ë°”ë¥¸ê³ ë”• í°íŠ¸ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤. ê¸°ë³¸ í°íŠ¸ë¡œ í‘œì‹œë©ë‹ˆë‹¤.")

except Exception as e:
    print(f"í°íŠ¸ ì„¤ì • ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
    print("ë‚˜ëˆ”ë°”ë¥¸ê³ ë”• í°íŠ¸ê°€ ì„¤ì¹˜ë˜ì–´ ìˆì§€ ì•Šê±°ë‚˜ ì„¤ì •ì— ë¬¸ì œê°€ ìˆìŠµë‹ˆë‹¤. ê¸°ë³¸ í°íŠ¸ë¡œ í‘œì‹œë©ë‹ˆë‹¤.")
    pass

# ë§ˆì´ë„ˆìŠ¤ ê¸°í˜¸ ê¹¨ì§ ë°©ì§€
plt.rcParams['axes.unicode_minus'] = False


# ğŸ“‹ ë¯¸ì…˜ 1: A/B í…ŒìŠ¤íŠ¸ ì‹ ë¢°êµ¬ê°„ ë¶„ì„ ğŸ§ª
print("="*20, "ë¯¸ì…˜ 1 ì‹œì‘", "="*20)
# ğŸ’¾ ë°ì´í„° ìƒì„± (ê·¸ëŒ€ë¡œ ì‹¤í–‰í•´ë³´ì„¸ìš”!)
# A/B í…ŒìŠ¤íŠ¸ ë°ì´í„° ìƒì„±
np.random.seed(404)

# Aê·¸ë£¹ (ê¸°ì¡´ ë””ìì¸): ì „í™˜ìœ¨ 12%
n_a = 500
conversion_rate_a = 0.12
conversions_a = np.random.binomial(1, conversion_rate_a, n_a)

# Bê·¸ë£¹ (ìƒˆ ë””ìì¸): ì „í™˜ìœ¨ 15%
n_b = 500
conversion_rate_b = 0.15
conversions_b = np.random.binomial(1, conversion_rate_b, n_b)

# ë°ì´í„°í”„ë ˆì„ ìƒì„±
ab_test_data = pd.DataFrame({
    'group': ['A']*n_a + ['B']*n_b,
    'user_id': range(1, n_a + n_b + 1),
    'converted': np.concatenate([conversions_a, conversions_b])
})

print("ğŸ§ª A/B í…ŒìŠ¤íŠ¸ ë°ì´í„°")
print(f"Aê·¸ë£¹ ì‚¬ìš©ì ìˆ˜: {n_a}ëª…")
print(f"Bê·¸ë£¹ ì‚¬ìš©ì ìˆ˜: {n_b}ëª…")
print("\nê·¸ë£¹ë³„ ì „í™˜ìœ¨:")
conversion_summary = ab_test_data.groupby('group')['converted'].agg(['count', 'sum', 'mean'])
conversion_summary.columns = ['ì´_ì‚¬ìš©ììˆ˜', 'ì „í™˜_ì‚¬ìš©ììˆ˜', 'ì „í™˜ìœ¨']
print(conversion_summary)


# âœ… ë¬¸ì œ 1-1: ê¸°ë³¸ í†µê³„ëŸ‰ ê³„ì‚°í•˜ê¸° ğŸ¯
print("\n--- ë¬¸ì œ 1-1 ---")
# TODO: p_a ê³„ì‚°í•˜ê¸° (Aê·¸ë£¹ ì „í™˜ìœ¨)
p_a = conversions_a.mean()

# TODO: p_b ê³„ì‚°í•˜ê¸° (Bê·¸ë£¹ ì „í™˜ìœ¨)
p_b = conversions_b.mean()

# TODO: Aê·¸ë£¹ í‘œì¤€ì˜¤ì°¨ ê³„ì‚°í•˜ê¸°
# íŒíŠ¸: ë¹„ìœ¨ì˜ í‘œì¤€ì˜¤ì°¨ ê³µì‹ì€ sqrt(p(1-p)/n)
se_a = np.sqrt(p_a * (1 - p_a) / n_a)

# TODO: Bê·¸ë£¹ í‘œì¤€ì˜¤ì°¨ ê³„ì‚°í•˜ê¸°
se_b = np.sqrt(p_b * (1 - p_b) / n_b)

print(f"Aê·¸ë£¹ ì „í™˜ìœ¨: {p_a:.3f} Â± {se_a:.3f}")
print(f"Bê·¸ë£¹ ì „í™˜ìœ¨: {p_b:.3f} Â± {se_b:.3f}")


# âœ… ë¬¸ì œ 1-2: Aê·¸ë£¹ ì‹ ë¢°êµ¬ê°„ ê³„ì‚°í•˜ê¸° ğŸ¯
print("\n--- ë¬¸ì œ 1-2 ---")
# TODO: 95% ì‹ ë¢°êµ¬ê°„ìš© zê°’ ì„¤ì •
z_score = 1.96

# TODO: Aê·¸ë£¹ ì‹ ë¢°êµ¬ê°„ í•˜í•œ ê³„ì‚°
ci_a_lower = p_a - z_score * se_a

# TODO: Aê·¸ë£¹ ì‹ ë¢°êµ¬ê°„ ìƒí•œ ê³„ì‚°
ci_a_upper = p_a + z_score * se_a

print(f"Aê·¸ë£¹ 95% ì‹ ë¢°êµ¬ê°„: [{ci_a_lower:.3f}, {ci_a_upper:.3f}]")


# âœ… å•é¡Œ 1-3: Bã‚°ãƒ«ãƒ¼ãƒ—ä¿¡é ¼åŒºé–“è¨ˆç®— ğŸ¯
print("\n--- å•é¡Œ 1-3 ---")
# TODO: Bê·¸ë£¹ ì‹ ë¢°êµ¬ê°„ í•˜í•œ ê³„ì‚°
ci_b_lower = p_b - z_score * se_b

# TODO: Bê·¸ë£¹ ì‹ ë¢°êµ¬ê°„ ìƒí•œ ê³„ì‚°
ci_b_upper = p_b + z_score * se_b

print(f"Bê·¸ë£¹ 95% ì‹ ë¢°êµ¬ê°„: [{ci_b_lower:.3f}, {ci_b_upper:.3f}]")


# âœ… å•é¡Œ 1-4: ê·¸ë£¹ë³„ ì‹ ë¢°êµ¬ê°„ ì‹œê°í™” ğŸ“Š
print("\n--- å•é¡Œ 1-4 (ã‚°ãƒ©ãƒ•ç”Ÿæˆ) ---")
# TODO: ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
plt.figure(figsize=(8, 6))

# TODO: ê·¸ë£¹ëª…ê³¼ ì „í™˜ìœ¨ ë¦¬ìŠ¤íŠ¸ ë§Œë“¤ê¸°
groups = ['Aê·¸ë£¹', 'Bê·¸ë£¹']
rates = [p_a, p_b]

# TODO: ì˜¤ì°¨ë§‰ëŒ€ í¬ê¸° ê³„ì‚° (z_score * í‘œì¤€ì˜¤ì°¨)
errors = [z_score * se_a, z_score * se_b]

# TODO: ë§‰ëŒ€ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
# ãƒ’ãƒ³ãƒˆ: plt.bar(x, height, yerr=ì˜¤ì°¨ë§‰ëŒ€, capsize=10, alpha=0.7)
plt.bar(groups, rates, yerr=errors, capsize=10, alpha=0.7, color=['skyblue', 'salmon'])

plt.ylabel('ì „í™˜ìœ¨')
plt.title('ê·¸ë£¹ë³„ ì „í™˜ìœ¨ê³¼ 95% ì‹ ë¢°êµ¬ê°„')
plt.ylim(0, 0.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# âœ… å•é¡Œ 1-5: ì „í™˜ìœ¨ ì°¨ì´ ê³„ì‚° ğŸ“Š
print("\n--- å•é¡Œ 1-5 ---")
# TODO: ì „í™˜ìœ¨ ì°¨ì´ ê³„ì‚° (B - A)
diff = p_b - p_a

# TODO: ì°¨ì´ì˜ í‘œì¤€ì˜¤ì°¨ ê³„ì‚°
# ãƒ’ãƒ³ãƒˆ: 2ã¤ã®ç‹¬ç«‹ã—ãŸæ¯”ç‡ã®å·®ã®æ¨™æº–èª¤å·®ã¯ sqrt(se_a^2 + se_b^2)
se_diff = np.sqrt(se_a**2 + se_b**2)

print(f"ì „í™˜ìœ¨ ì°¨ì´ (B-A): {diff:.3f}")
print(f"ì°¨ì´ì˜ í‘œì¤€ì˜¤ì°¨: {se_diff:.3f}")


# âœ… å•é¡Œ 1-6: ì°¨ì´ì˜ ì‹ ë¢°êµ¬ê°„ ê³„ì‚° ğŸ¯
print("\n--- å•é¡Œ 1-6 ---")
# TODO: ì°¨ì´ì˜ ì‹ ë¢°êµ¬ê°„ í•˜í•œ ê³„ì‚°
ci_diff_lower = diff - 1.96 * se_diff

# TODO: ì°¨ì´ì˜ ì‹ ë¢°êµ¬ê°„ ìƒí•œ ê³„ì‚°
ci_diff_upper = diff + 1.96 * se_diff

print(f"ì „í™˜ìœ¨ ì°¨ì´ì˜ 95% ì‹ ë¢°êµ¬ê°„: [{ci_diff_lower:.3f}, {ci_diff_upper:.3f}]")

# TODO: 0ì´ í¬í•¨ë˜ëŠ”ì§€ í™•ì¸í•˜ê¸°
if ci_diff_lower > 0:
    print("ğŸ‰ Bê°€ Aë³´ë‹¤ í™•ì‹¤íˆ ì¢‹ìŠµë‹ˆë‹¤!")
elif ci_diff_upper < 0:
    print("ğŸ˜¢ Aê°€ Bë³´ë‹¤ í™•ì‹¤íˆ ì¢‹ìŠµë‹ˆë‹¤.")
else:
    print("ğŸ¤” í™•ì‹¤í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")


# âœ… å•é¡Œ 1-7: ê²½ì˜ì§„ ë³´ê³ ì„œ ì‘ì„± ğŸ’¼
print("\n--- å•é¡Œ 1-7 ---")
# TODO: ë³´ê³ ì„œ ì™„ì„±í•˜ê¸°
print("ğŸ“‹ ê²½ì˜ì§„ ë³´ê³ ì„œ")
print("="*50)
print(f"ğŸ” ë¶„ì„ ê²°ê³¼:")
print(f"   â€¢ Aê·¸ë£¹ ì „í™˜ìœ¨: {p_a:.1%}")
print(f"   â€¢ Bê·¸ë£¹ ì „í™˜ìœ¨: {p_b:.1%}")
print(f"   â€¢ ê°œì„  íš¨ê³¼: {diff:.1%}p")
print()
print(f"ğŸ“Š í†µê³„ì  ì‹ ë¢°ë„:")
print(f"   â€¢ ê°œì„  íš¨ê³¼ ë²”ìœ„: {ci_diff_lower:.1%}p ~ {ci_diff_upper:.1%}p (95% ì‹ ë¢°ìˆ˜ì¤€)")
print()
print("ğŸ’¡ ì˜ì‚¬ê²°ì •:")

# TODO: ifë¬¸ì„ ì‚¬ìš©í•´ì„œ ê¶Œì¥ì‚¬í•­ ì‘ì„±í•˜ê¸°
# ãƒ’ãƒ³ãƒˆ: ci_diff_lower > 0ã®å ´åˆã€æ–°ã—ã„ãƒ‡ã‚¶ã‚¤ãƒ³ã®å°å…¥ã‚’æ¨å¥¨
if ci_diff_lower > 0:
    print("   âœ… ìƒˆ ë””ìì¸(B) ë„ì…ì„ ê¶Œì¥í•©ë‹ˆë‹¤! B ë””ìì¸ì´ ê¸°ì¡´ë³´ë‹¤ ì „í™˜ìœ¨ì„ ë†’ì¼ ê°€ëŠ¥ì„±ì´ ë†’ìŠµë‹ˆë‹¤.")
else:
    print("   âŒ ë‘ ë””ìì¸ ê°„ì˜ íš¨ê³¼ ì°¨ì´ê°€ í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. ì¶”ê°€ ë°ì´í„° ìˆ˜ì§‘ ë˜ëŠ” ë‹¤ë¥¸ í…ŒìŠ¤íŠ¸ë¥¼ ê³ ë ¤í•´ì•¼ í•©ë‹ˆë‹¤.")

print("\n", "="*20, "ë¯¸ì…˜ 1 ì¢…ë£Œ", "="*20, "\n")


# ğŸ“‹ ãƒŸãƒƒã‚·ãƒ§ãƒ³ 2: å¹´é½¢å±¤åˆ¥è³¼è²·åˆ†æ (t-æ¤œå®šãƒã‚¹ã‚¿ãƒ¼) ğŸ‘¥
print("="*20, "ë¯¸ì…˜ 2 ì‹œì‘", "="*20)
# ğŸ’¾ ãƒ‡ãƒ¼ã‚¿ç”Ÿæˆ
# ã‚ªãƒ³ãƒ©ã‚¤ãƒ³ã‚·ãƒ§ãƒƒãƒ”ãƒ³ã‚°ãƒ¢ãƒ¼ãƒ«ã®è³¼è²·ãƒ‡ãƒ¼ã‚¿
np.random.seed(789)
ages = np.random.randint(20, 60, 100)
purchase_amounts = 50 + 2*ages + np.random.normal(0, 20, 100)

print("ğŸ“Š è³¼è²·ãƒ‡ãƒ¼ã‚¿ãƒ—ãƒ¬ãƒ“ãƒ¥ãƒ¼")
print(f"ì—°ë ¹ ë²”ìœ„: {ages.min()}ì„¸ ~ {ages.max()}ì„¸")
print(f"êµ¬ë§¤ê¸ˆì•¡ ë²”ìœ„: {purchase_amounts.min():.0f}ë§Œì› ~ {purchase_amounts.max():.0f}ë§Œì›")
print(f"í‰ê·  êµ¬ë§¤ê¸ˆì•¡: {purchase_amounts.mean():.1f}ë§Œì›")


# âœ… å•é¡Œ 2-1: åŸºæœ¬çµ±è¨ˆé‡è¨ˆç®— ğŸ¯
print("\n--- å•é¡Œ 2-1 ---")
# TODO: í‘œë³¸ í‰ê·  ê³„ì‚°
sample_mean = purchase_amounts.mean()

# TODO: í‘œë³¸ í‘œì¤€í¸ì°¨ ê³„ì‚°
sample_std = purchase_amounts.std(ddof=1)

# TODO: í‘œë³¸ í¬ê¸° ê³„ì‚°
n = len(purchase_amounts)

# TODO: ë¹„êµí•  ê¸°ì¤€ê°’ ì„¤ì • (ì—…ê³„ ê¸°ì¤€ 150ë§Œì›)
standard_value = 150

print(f"í‘œë³¸ í‰ê· : {sample_mean:.2f}ë§Œì›")
print(f"í‘œë³¸ í‘œì¤€í¸ì°¨: {sample_std:.2f}ë§Œì›")
print(f"í‘œë³¸ í¬ê¸°: {n}ëª…")
print(f"ë¹„êµ ê¸°ì¤€ê°’: {standard_value}ë§Œì›")


# âœ… å•é¡Œ 2-2: t-çµ±è¨ˆé‡ç›´æ¥è¨ˆç®— ğŸ§®
print("\n--- å•é¡Œ 2-2 ---")
# TODO: t-í†µê³„ëŸ‰ ê³µì‹ ì‚¬ìš©í•˜ê¸°
# ãƒ’ãƒ³ãƒˆ: t = (æ¨™æœ¬å¹³å‡ - åŸºæº–å€¤) / (æ¨™æœ¬æ¨™æº–åå·® / sqrt(æ¨™æœ¬ã‚µã‚¤ã‚º))
t_stat = (sample_mean - standard_value) / (sample_std / np.sqrt(n))

print(f"ì§ì ‘ ê³„ì‚°í•œ t-í†µê³„ëŸ‰: {t_stat:.3f}")

# ë¹„êµë¥¼ ìœ„í•´ scipyë¡œë„ ê³„ì‚°í•´ë³´ê¸°
from scipy.stats import ttest_1samp
t_stat_scipy, p_value = ttest_1samp(purchase_amounts, standard_value)
print(f"scipyë¡œ ê³„ì‚°í•œ t-í†µê³„ëŸ‰: {t_stat_scipy:.3f}")
print(f"p-value: {p_value:.4f}")


# âœ… å•é¡Œ 2-3: çµæœè§£é‡ˆ ğŸ¯
print("\n--- å•é¡Œ 2-3 ---")
# TODO: ìœ ì˜ìˆ˜ì¤€ ì„¤ì •
alpha = 0.05

print("ğŸ¯ ê²°ê³¼ í•´ì„")

# TODO: p-valueì™€ alpha ë¹„êµí•˜ê¸°
if p_value < alpha:
    print(f"âœ… p-value ({p_value:.4f}) < Î± ({alpha})")
    print("â†’ ê·€ë¬´ê°€ì„¤ ê¸°ê°! ìš°ë¦¬ ì‡¼í•‘ëª° ê³ ê°ì˜ í‰ê·  êµ¬ë§¤ê¸ˆì•¡ì€ ì—…ê³„ ê¸°ì¤€ê³¼ ìœ ì˜í•˜ê²Œ ë‹¤ë¦…ë‹ˆë‹¤.")

    # TODO: í‰ê· ì´ ë” í°ì§€ ì‘ì€ì§€ í™•ì¸í•˜ê¸°
    if sample_mean > standard_value:
        print("â†’ ìš°ë¦¬ ê³ ê°ë“¤ì´ ì—…ê³„ í‰ê· ë³´ë‹¤ ë” ë§ì´ êµ¬ë§¤í•©ë‹ˆë‹¤! ğŸ‰")
    else:
        print("â†’ ìš°ë¦¬ ê³ ê°ë“¤ì´ ì—…ê³„ í‰ê· ë³´ë‹¤ ì ê²Œ êµ¬ë§¤í•©ë‹ˆë‹¤. ğŸ˜…")
else:
    print(f"âŒ p-value ({p_value:.4f}) â‰¥ Î± ({alpha})")
    print("â†’ ìš°ë¦¬ ì‡¼í•‘ëª° ê³ ê°ì˜ í‰ê·  êµ¬ë§¤ê¸ˆì•¡ì€ ì—…ê³„ ê¸°ì¤€ê³¼ ìœ ì˜í•œ ì°¨ì´ê°€ ì—†ìŠµë‹ˆë‹¤.")


# âœ… å•é¡Œ 2-4: å¹´é½¢å±¤åˆ¥ã‚°ãƒ«ãƒ¼ãƒ—åˆ†ã‘ ğŸ‘¥
print("\n--- å•é¡Œ 2-4 ---")
# TODO: ì Šì€ì¸µ ë°ì´í„° í•„í„°ë§ (20-39ì„¸)
young_mask = (ages >= 20) & (ages <= 39)
group1_purchases = purchase_amounts[young_mask]

# TODO: ì¤‘ë…„ì¸µ ë°ì´í„° í•„í„°ë§ (40-59ì„¸)
middle_mask = (ages >= 40) & (ages <= 59)
group2_purchases = purchase_amounts[middle_mask]

print("ğŸ‘¥ ê·¸ë£¹ ë‚˜ëˆ„ê¸° ê²°ê³¼")
print(f"ì Šì€ì¸µ (20-39ì„¸): {len(group1_purchases)}ëª…")
print(f"ì¤‘ë…„ì¸µ (40-59ì„¸): {len(group2_purchases)}ëª…")


# âœ… å•é¡Œ 2-5: ã‚°ãƒ«ãƒ¼ãƒ—åˆ¥è¨˜è¿°çµ±è¨ˆé‡è¨ˆç®— ğŸ“Š
print("\n--- å•é¡Œ 2-5 ---")
# TODO: ê° ê·¸ë£¹ì˜ í‰ê·  ê³„ì‚°
group1_mean = group1_purchases.mean()
group2_mean = group2_purchases.mean()

# TODO: ê° ê·¸ë£¹ì˜ í‘œì¤€í¸ì°¨ ê³„ì‚°
group1_std = group1_purchases.std()
group2_std = group2_purchases.std()

print("ğŸ“Š ê·¸ë£¹ë³„ ê¸°ìˆ í†µê³„")
print(f"ì Šì€ì¸µ í‰ê· : {group1_mean:.1f}ë§Œì› (í‘œì¤€í¸ì°¨: {group1_std:.1f})")
print(f"ì¤‘ë…„ì¸µ í‰ê· : {group2_mean:.1f}ë§Œì› (í‘œì¤€í¸ì°¨: {group2_std:.1f})")

# TODO: í‰ê·  ì°¨ì´ ê³„ì‚°
diff = group2_mean - group1_mean
print(f"í‰ê·  ì°¨ì´: {diff:.1f}ë§Œì›")


# âœ… å•é¡Œ 2-6: ç‹¬ç«‹æ¨™æœ¬t-æ¤œå®šå®Ÿè¡Œ ğŸ§ª
print("\n--- å•é¡Œ 2-6 ---")
from scipy.stats import ttest_ind

# TODO: ë…ë¦½í‘œë³¸ t-ê²€ì • ì‹¤í–‰
t_stat, p_value = ttest_ind(group1_purchases, group2_purchases)

print("ğŸ§® t-ê²€ì • ê²°ê³¼")
print(f"t-í†µê³„ëŸ‰: {t_stat:.3f}")
print(f"p-value: {p_value:.4f}")

# TODO: ê²°ê³¼ í•´ì„í•˜ê¸°
alpha = 0.05
print("ğŸ¯ ãƒ“ì¦ˆãƒã‚¹è§£é‡ˆ")

if p_value < alpha:
    print("âœ… í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•œ ì°¨ì´ê°€ ìˆìŠµë‹ˆë‹¤!")

    if group2_mean > group1_mean:
        print("â†’ ì¤‘ë…„ì¸µì´ ì Šì€ì¸µë³´ë‹¤ ë” ë§ì´ êµ¬ë§¤í•©ë‹ˆë‹¤!")
    else:
        print("â†’ ì Šì€ì¸µì´ ì¤‘ë…„ì¸µë³´ë‹¤ ë” ë§ì´ êµ¬ë§¤í•©ë‹ˆë‹¤!")
else:
    print("âŒ ì—°ë ¹ëŒ€ë³„ êµ¬ë§¤ê¸ˆì•¡ì— í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•œ ì°¨ì´ê°€ ì—†ìŠµë‹ˆë‹¤.")


# âœ… å•é¡Œ 2-7: ç­‰åˆ†æ•£æ€§æ¤œå®š ğŸ”
print("\n--- å•é¡Œ 2-7 ---")
from scipy.stats import levene

# TODO: ë¶„ì‚° ì§ì ‘ ë¹„êµí•˜ê¸°
var1 = group1_purchases.var(ddof=1)
var2 = group2_purchases.var(ddof=1)

print("ğŸ“Š åˆ†æ•£æ¯”è¼ƒ")
print(f"ì Šì€ì¸µ ë¶„ì‚°: {var1:.1f}")
print(f"ì¤‘ë…„ì¸µ ë¶„ì‚°: {var2:.1f}")

# TODO: Levene ê²€ì • ì‹¤í–‰í•˜ê¸°
levene_stat, levene_p = levene(group1_purchases, group2_purchases)

print("ğŸ§ª Levene ë“±ë¶„ì‚°ì„± ê²€ì •")
print(f"p-value: {levene_p:.4f}")

# TODO: ë“±ë¶„ì‚°ì„± íŒë‹¨í•˜ê¸°
if levene_p >= 0.05:
    print("âœ… ë“±ë¶„ì‚°ì„± ê°€ì •ì„ ë§Œì¡±í•©ë‹ˆë‹¤. (ë‘ ê·¸ë£¹ì˜ ë¶„ì‚°ì´ ë¹„ìŠ·í•©ë‹ˆë‹¤)")
else:
    print("âŒ ë“±ë¶„ì‚°ì„± ê°€ì •ì„ ìœ„ë°°í•©ë‹ˆë‹¤. (ë‘ ê·¸ë£¹ì˜ ë¶„ì‚°ì´ ë‹¤ë¦…ë‹ˆë‹¤)")
    print("   (ã“ã®å ´åˆã€ttest_ind(..., equal_var=False)ã‚ªãƒ—ã‚·ãƒ§ãƒ³ã‚’ä½¿ç”¨ã™ã‚‹ã¨ã‚ˆã‚Šæ­£ç¢ºã§ã™ã€‚)")

print("\n", "="*20, "ãƒŸãƒƒã‚·ãƒ§ãƒ³ 2 çµ‚äº†", "="*20, "\n")


# ğŸ“‹ ãƒŸãƒƒã‚·ãƒ§ãƒ³ 3: å¾“æ¥­å“¡æº€è¶³åº¦ç·åˆåˆ†æ ğŸ¢
print("="*20, "ë¯¸ì…˜ 3 ì‹œì‘", "="*20)
# ğŸ’¾ ãƒ‡ãƒ¼ã‚¿ç”Ÿæˆ
# å¾“æ¥­å“¡æº€è¶³åº¦ç·åˆãƒ‡ãƒ¼ã‚¿ç”Ÿæˆ
np.random.seed(2024)
n = 250

data_hr = {
    'ì—°ë´‰': np.random.lognormal(np.log(50000), 0.3, n),
    'ê·¼ë¬´ì—°ìˆ˜': np.random.exponential(3, n),
    'ë¶€ì„œ': np.random.choice(['ì˜ì—…', 'ê°œë°œ', 'ë§ˆì¼€íŒ…', 'ì¸ì‚¬'], n, p=[0.3, 0.4, 0.2, 0.1]),
    'êµìœ¡ì‹œê°„': np.random.gamma(2, 10, n),
    'ì§ê¸‰': np.random.choice(['ì‚¬ì›', 'ëŒ€ë¦¬', 'ê³¼ì¥', 'ë¶€ì¥'], n, p=[0.4, 0.3, 0.2, 0.1])
}

df_hr = pd.DataFrame(data_hr)

# æº€è¶³åº¦ã‚¹ã‚³ã‚¢ (1-10ç‚¹) ç”Ÿæˆ
satisfaction_base = (
    df_hr['ì—°ë´‰']/10000 * 0.8 +
    df_hr['ê·¼ë¬´ì—°ìˆ˜'] * 0.3 +
    df_hr['êµìœ¡ì‹œê°„'] * 0.1
)

dept_effect = df_hr['ë¶€ì„œ'].map({'ì˜ì—…': -0.5, 'ê°œë°œ': 0.8, 'ë§ˆì¼€íŒ…': 0.3, 'ì¸ì‚¬': 0.0})
rank_effect = df_hr['ì§ê¸‰'].map({'ì‚¬ì›': 0, 'ëŒ€ë¦¬': 0.5, 'ê³¼ì¥': 1.0, 'ë¶€ì¥': 1.5})

satisfaction = satisfaction_base + dept_effect + rank_effect + np.random.normal(0, 0.8, n)
satisfaction = np.clip(satisfaction, 1, 10)

# ì´ì§ ì˜í–¥
turnover_prob = 1 / (1 + np.exp(satisfaction - 5))
turnover_intent = np.random.binomial(1, turnover_prob, n)

df_hr['ë§Œì¡±ë„'] = satisfaction.round(1)
df_hr['ì´ì§ì˜í–¥'] = turnover_intent

print("ğŸ“Š å¾“æ¥­å“¡æº€è¶³åº¦ãƒ‡ãƒ¼ã‚¿ (æœ€åˆã®5è¡Œ)")
print(df_hr.head())


# âœ… å•é¡Œ 3-1: ç›¸é–¢ä¿‚æ•°è¡Œåˆ—ä½œæˆ ğŸ”
print("\n--- å•é¡Œ 3-1 ---")
# TODO: ìˆ«ìí˜• ë³€ìˆ˜ë“¤ë§Œ ì„ ë³„í•˜ê¸°
numeric_cols = ['ì—°ë´‰', 'ê·¼ë¬´ì—°ìˆ˜', 'êµìœ¡ì‹œê°„', 'ë§Œì¡±ë„', 'ì´ì§ì˜í–¥']
correlation_data = df_hr[numeric_cols]

# TODO: ìƒê´€ê³„ìˆ˜ ë§¤íŠ¸ë¦­ìŠ¤ ê³„ì‚°í•˜ê¸°
correlation_matrix = correlation_data.corr()

print("ğŸ“Š ç›¸é–¢ä¿‚æ•°è¡Œåˆ—")
print(correlation_matrix.round(3))


# âœ… å•é¡Œ 3-2: æº€è¶³åº¦ã¨ã®ç›¸é–¢é–¢ä¿‚åˆ†æ ğŸ¯
print("\n--- å•é¡Œ 3-2 ---")
# TODO: ë§Œì¡±ë„ì™€ì˜ ìƒê´€ê´€ê³„ë§Œ ì¶”ì¶œí•˜ê¸°
satisfaction_corr = correlation_matrix['ë§Œì¡±ë„'].sort_values(ascending=False)

print("ğŸ¯ æº€è¶³åº¦ã¨ã®ç›¸é–¢é–¢ä¿‚ (é«˜ã„é †)")

# TODO: forë¬¸ìœ¼ë¡œ ê° ë³€ìˆ˜ë³„ ìƒê´€ê´€ê³„ ì¶œë ¥í•˜ê¸°
for var, corr in satisfaction_corr.items():
    if var != 'ë§Œì¡±ë„':  # ë§Œì¡±ë„ ìê¸° ìì‹  ì œì™¸
        # TODO: ìƒê´€ê´€ê³„ ê°•ë„ íŒë‹¨í•˜ê¸°
        if abs(corr) > 0.5:
            strength = "ê°•í•œ"
        elif abs(corr) > 0.3:
            strength = "ì¤‘ê°„"
        else:
            strength = "ì•½í•œ"

        direction = "ì–‘ì˜" if corr > 0 else "ìŒì˜"
        print(f"â€¢ {var}: {corr:.3f} ({strength} {direction} ìƒê´€ê´€ê³„)")


# âœ… å•é¡Œ 3-3: ç›¸é–¢é–¢ä¿‚ãƒ’ãƒ¼ãƒˆãƒãƒƒãƒ—æç”» ğŸ“ˆ
print("\n--- å•é¡Œ 3-3 (ã‚°ãƒ©ãƒ•ç”Ÿæˆ) ---")
# TODO: íˆíŠ¸ë§µ ê·¸ë¦¬ê¸°
plt.figure(figsize=(10, 8))

# TODO: seaborn heatmap ì‚¬ìš©í•˜ê¸°
# ãƒ’ãƒ³ãƒˆ: sns.heatmap(data, annot=True, cmap='RdBu_r', center=0, fmt='.3f')
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu_r', center=0, fmt='.3f')

plt.title('ğŸ“Š å¤‰æ•°é–“ã®ç›¸é–¢é–¢ä¿‚ãƒ’ãƒ¼ãƒˆãƒãƒƒãƒ—')
plt.tight_layout()
plt.show()


# âœ… å•é¡Œ 3-4: éƒ¨ç½²åˆ¥æº€è¶³åº¦æ¯”è¼ƒ ğŸ¢
print("\n--- å•é¡Œ 3-4 ---")
# TODO: ë¶€ì„œë³„ ê·¸ë£¹í™”í•´ì„œ í‰ê·  ê³„ì‚°í•˜ê¸°
dept_summary = df_hr.groupby('ë¶€ì„œ')['ë§Œì¡±ë„'].mean()

print("ğŸ¢ ë¶€ì„œë³„ í‰ê·  ë§Œì¡±ë„")
print(dept_summary.round(2))

# TODO: ë¶€ì„œë³„ ì´ì§ì˜í–¥ë„ ê³„ì‚°í•˜ê¸°
dept_turnover = df_hr.groupby('ë¶€ì„œ')['ì´ì§ì˜í–¥'].mean()

print("\nğŸ¢ ë¶€ì„œë³„ ì´ì§ì˜í–¥")
print(dept_turnover.round(3))

# TODO: forë¬¸ìœ¼ë¡œ ë¶€ì„œë³„ ìƒì„¸ ë¶„ì„ ì¶œë ¥í•˜ê¸°
print("\nğŸ“‹ ë¶€ì„œë³„ ìƒì„¸ ë¶„ì„")
for dept in df_hr['ë¶€ì„œ'].unique():
    # TODO: í•´ë‹¹ ë¶€ì„œ ë°ì´í„°ë§Œ í•„í„°ë§í•˜ê¸°
    dept_data = df_hr[df_hr['ë¶€ì„œ'] == dept]

    satisfaction_avg = dept_data['ë§Œì¡±ë„'].mean()
    turnover_avg = dept_data['ì´ì§ì˜í–¥'].mean()

    print(f"{dept}ë¶€: ë§Œì¡±ë„ {satisfaction_avg:.1f}ì , ì´ì§ì˜í–¥ {turnover_avg:.1%}")


# âœ… å•é¡Œ 3-5: ãƒ€ãƒŸãƒ¼å¤‰æ•°ä½œæˆ ğŸ”§
print("\n--- å•é¡Œ 3-5 ---")
# TODO: ë¶€ì„œë¥¼ ë”ë¯¸ ë³€ìˆ˜ë¡œ ë³€í™˜í•˜ê¸°
dept_dummies = pd.get_dummies(df_hr['ë¶€ì„œ'], prefix='ë¶€ì„œ')

# TODO: ì§ê¸‰ì„ ë”ë¯¸ ë³€ìˆ˜ë¡œ ë³€í™˜í•˜ê¸°
rank_dummies = pd.get_dummies(df_hr['ì§ê¸‰'], prefix='ì§ê¸‰')

# TODO: ê¸°ì¤€ ì¹´í…Œê³ ë¦¬ ì œê±°í•˜ê¸° (ë‹¤ì¤‘ê³µì„ ì„± ë°©ì§€)
dept_dummies = dept_dummies.drop('ë¶€ì„œ_ì¸ì‚¬', axis=1)
rank_dummies = rank_dummies.drop('ì§ê¸‰_ì‚¬ì›', axis=1)

print("ìƒì„±ëœ ë”ë¯¸ ë³€ìˆ˜ë“¤:")
print("ë¶€ì„œ:", list(dept_dummies.columns))
print("ì§ê¸‰:", list(rank_dummies.columns))


# âœ… å•é¡Œ 3-6: å›å¸°åˆ†æç”¨ãƒ‡ãƒ¼ã‚¿æº–å‚™ ğŸ”§
print("\n--- å•é¡Œ 3-6 ---")
# TODO: ë…ë¦½ë³€ìˆ˜ ë°ì´í„°í”„ë ˆì„ ë§Œë“¤ê¸°
X_vars = pd.concat([df_hr[['ì—°ë´‰', 'ê·¼ë¬´ì—°ìˆ˜', 'êµìœ¡ì‹œê°„']], dept_dummies, rank_dummies], axis=1)

# TODO: ì¢…ì†ë³€ìˆ˜ ì„¤ì •í•˜ê¸°
y_var = df_hr['ë§Œì¡±ë„']

print("ğŸ”§ å›å¸°åˆ†æå¤‰æ•°æ§‹æˆ")
print(f"ë…ë¦½ë³€ìˆ˜ ê°œìˆ˜: {len(X_vars.columns)}ê°œ")
print(f"ë…ë¦½ë³€ìˆ˜: {list(X_vars.columns)}")
print(f"ì¢…ì†ë³€ìˆ˜: ë§Œì¡±ë„")


# âœ… å•é¡Œ 3-7: å›å¸°åˆ†æå®Ÿè¡Œ ğŸ“Š
print("\n--- å•é¡Œ 3-7 ---")
# TODO: íšŒê·€ëª¨ë¸ ìƒì„± ë° í•™ìŠµí•˜ê¸°
model = LinearRegression()
model.fit(X_vars, y_var)

# TODO: ì˜ˆì¸¡ê°’ ê³„ì‚°í•˜ê¸°
y_pred = model.predict(X_vars)

# TODO: RÂ² ê³„ì‚°í•˜ê¸°
r2 = r2_score(y_var, y_pred)

print("ğŸ“Š é‡å›å¸°åˆ†æçµæœ")
print(f"RÂ² (ì„¤ëª…ë ¥): {r2:.3f} ({r2*100:.1f}%)")
print("â†’ ë…ë¦½ë³€ìˆ˜ë“¤ì´ ë§Œì¡±ë„ ë³€ë™ì˜ ì•½ 98.7%ë¥¼ ì„¤ëª…í•©ë‹ˆë‹¤.")


# âœ… å•é¡Œ 3-8: å›å¸°ä¿‚æ•°è§£é‡ˆ ğŸ”
print("\n--- å•é¡Œ 3-8 ---")
# TODO: íšŒê·€ê³„ìˆ˜ë¥¼ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ì •ë¦¬í•˜ê¸°
coefficients = pd.DataFrame({
    'ë³€ìˆ˜': X_vars.columns,
    'ê³„ìˆ˜': model.coef_,
    'ì ˆëŒ“ê°’': np.abs(model.coef_)
})

# TODO: ì ˆëŒ“ê°’ ê¸°ì¤€ìœ¼ë¡œ ì •ë ¬í•˜ê¸°
coefficients = coefficients.sort_values('ì ˆëŒ“ê°’', ascending=False)

print("ğŸ” å›å¸°ä¿‚æ•°åˆ†æ (å½±éŸ¿åŠ›ã®å¤§ãã„é †)")
print(coefficients.round(3).head(5))  # ìƒìœ„ 5ê°œë§Œ ì¶œë ¥

# TODO: ê°€ì¥ ì˜í–¥ë ¥ì´ í° ë³€ìˆ˜ ì°¾ê¸°
most_important = coefficients.iloc[0]
print(f"\nê°€ì¥ ì˜í–¥ë ¥ì´ í° ë³€ìˆ˜: {most_important['ë³€ìˆ˜']} (ê³„ìˆ˜: {most_important['ê³„ìˆ˜']:.3f})")


# âœ… å•é¡Œ 3-9: ãƒ­ã‚¸ã‚¹ãƒ†ã‚£ãƒƒã‚¯å›å¸°ãƒ‡ãƒ¼ã‚¿æº–å‚™ ğŸ¯
print("\n--- å•é¡Œ 3-9 ---")
# TODO: ë¡œì§€ìŠ¤í‹± íšŒê·€ìš© ë…ë¦½ë³€ìˆ˜ì™€ ì¢…ì†ë³€ìˆ˜ ì„¤ì •í•˜ê¸°
X_logistic = X_vars # ìœ„ì—ì„œ ë§Œë“  ê²ƒê³¼ ë™ì¼
y_logistic = df_hr['ì´ì§ì˜í–¥']

# TODO: í˜„ì¬ ì´ì§ ì˜í–¥ í˜„í™© íŒŒì•…í•˜ê¸°
turnover_summary = df_hr['ì´ì§ì˜í–¥'].value_counts()
turnover_rate = df_hr['ì´ì§ì˜í–¥'].mean()

print("ğŸ“Š ç¾åœ¨ã®é›¢è·æ„å‘ç¾çŠ¶")
print(f"ì´ì§ ì˜í–¥ ìˆìŒ: {turnover_summary[1]}ëª… ({turnover_rate:.1%})")
print(f"ì´ì§ ì˜í–¥ ì—†ìŒ: {turnover_summary[0]}ëª… ({1-turnover_rate:.1%})")


# âœ… å•é¡Œ 3-10: ãƒ­ã‚¸ã‚¹ãƒ†ã‚£ãƒƒã‚¯å›å¸°ãƒ¢ãƒ‡ãƒ«å­¦ç¿’ ğŸ¤–
print("\n--- å•é¡Œ 3-10 ---")
# TODO: ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ ìƒì„±í•˜ê¸°
logistic_model = LogisticRegression(random_state=42, max_iter=1000)

# TODO: ëª¨ë¸ í•™ìŠµí•˜ê¸°
logistic_model.fit(X_logistic, y_logistic)

# TODO: ì´ì§ í™•ë¥  ì˜ˆì¸¡í•˜ê¸°
y_pred_proba = logistic_model.predict_proba(X_logistic)[:, 1]

# TODO: ì´ì§ ì—¬ë¶€ ì˜ˆì¸¡í•˜ê¸° (0 ë˜ëŠ” 1)
y_pred_class = logistic_model.predict(X_logistic)

# TODO: ì •í™•ë„ ê³„ì‚°í•˜ê¸°
accuracy = (y_pred_class == y_logistic).mean()

print("ğŸ¤– ãƒ­ã‚¸ã‚¹ãƒ†ã‚£ãƒƒã‚¯å›å¸°ãƒ¢ãƒ‡ãƒ«æ€§èƒ½")
print(f"ì •í™•ë„: {accuracy:.3f}")


# âœ… å•é¡Œ 3-11: æ··åŒè¡Œåˆ—ä½œæˆ ğŸ“‹
print("\n--- å•é¡Œ 3-11 ---")
# TODO: í˜¼ë™ í–‰ë ¬ ê³„ì‚°í•˜ê¸°
cm = confusion_matrix(y_logistic, y_pred_class)

print("ğŸ“‹ æ··åŒè¡Œåˆ—")
print("ì‹¤ì œ â†“ | ì˜ˆì¸¡ â†’   ì´ì§ì•ˆí•¨  ì´ì§í•¨")
print("-----------------------------------")
print(f"ì´ì§ ì•ˆí•¨        {cm[0,0]:>7d}    {cm[0,1]:>5d}")
print(f"ì´ì§ í•¨          {cm[1,0]:>7d}    {cm[1,1]:>5d}")

# TODO: ì •ë°€ë„ì™€ ì¬í˜„ìœ¨ ê³„ì‚°í•˜ê¸°
# ç²¾å¯†ë„ = TP / (TP + FP)
precision = cm[1,1] / (cm[1,1] + cm[0,1]) if (cm[1,1] + cm[0,1]) > 0 else 0

# å†ç¾ç‡ = TP / (TP + FN)
recall = cm[1,1] / (cm[1,1] + cm[1,0]) if (cm[1,1] + cm[1,0]) > 0 else 0

print(f"\nì •ë°€ë„ (Precision): {precision:.3f} (ëª¨ë¸ì´ 'ì´ì§í•œë‹¤'ê³  ì˜ˆì¸¡í•œ ê²ƒ ì¤‘ ì‹¤ì œ ì´ì§í•œ ë¹„ìœ¨)")
print(f"ì¬í˜„ìœ¨ (Recall): {recall:.3f} (ì‹¤ì œ ì´ì§ ëŒ€ìƒì ì¤‘ ëª¨ë¸ì´ 'ì´ì§í•œë‹¤'ê³  ì°¾ì•„ë‚¸ ë¹„ìœ¨)")


# âœ… å•é¡Œ 3-12: ã‚ªãƒƒã‚ºæ¯”è¨ˆç®— ğŸ“ˆ
print("\n--- å•é¡Œ 3-12 ---")
# TODO: ì˜¤ì¦ˆë¹„ ê³„ì‚°í•˜ê¸°
# ãƒ’ãƒ³ãƒˆ: ã‚ªãƒƒã‚ºæ¯” = exp(å›å¸°ä¿‚æ•°)
odds_ratios = np.exp(logistic_model.coef_[0])

# TODO: ì˜¤ì¦ˆë¹„ ë¶„ì„ ë°ì´í„°í”„ë ˆì„ ë§Œë“¤ê¸°
odds_analysis = pd.DataFrame({
    'ë³€ìˆ˜': X_logistic.columns,
    'ê³„ìˆ˜': logistic_model.coef_[0],
    'ì˜¤ì¦ˆë¹„': odds_ratios,
    'ìœ„í—˜ë„ë³€í™”(%)': [(or_val - 1) * 100 for or_val in odds_ratios]
})

# TODO: ìœ„í—˜ë„ ë³€í™”ê°€ í° ìˆœìœ¼ë¡œ ì •ë ¬í•˜ê¸°
odds_analysis = odds_analysis.sort_values('ìœ„í—˜ë„ë³€í™”(%)', key=abs, ascending=False)

print("ğŸ¯ é›¢è·ãƒªã‚¹ã‚¯è¦å› åˆ†æ (ä¸Šä½5ã¤)")
print(odds_analysis.round(3).head())


# âœ… å•é¡Œ 3-13: é«˜ãƒªã‚¹ã‚¯ç¾¤è­˜åˆ¥ ğŸš¨
print("\n--- å•é¡Œ 3-13 ---")
# TODO: ê³ ìœ„í—˜êµ° ê¸°ì¤€ ì„¤ì •í•˜ê¸° (70% ì´ìƒ)
high_risk_threshold = 0.7

# TODO: ê³ ìœ„í—˜êµ° ì§ì› í•„í„°ë§í•˜ê¸°
high_risk_mask = y_pred_proba > high_risk_threshold
high_risk_employees = df_hr[high_risk_mask].copy()
high_risk_employees['ì˜ˆì¸¡ì´ì§í™•ë¥ '] = y_pred_proba[high_risk_mask]


print(f"ğŸ¯ ì´ì§ ê³ ìœ„í—˜êµ° ì‹ë³„")
print(f"ì´ì§ í™•ë¥  {high_risk_threshold:.0%} ì´ìƒ ì§ì›: {len(high_risk_employees)}ëª…")

if len(high_risk_employees) > 0:
    print("ê³ ìœ„í—˜êµ° íŠ¹ì§•:")
    # TODO: ê³ ìœ„í—˜êµ°ì˜ í‰ê·  ë§Œì¡±ë„ ê³„ì‚°í•˜ê¸°
    avg_satisfaction = high_risk_employees['ë§Œì¡±ë„'].mean()

    # TODO: ê³ ìœ„í—˜êµ°ì˜ í‰ê·  ì—°ë´‰ ê³„ì‚°í•˜ê¸°
    avg_salary = high_risk_employees['ì—°ë´‰'].mean()

    print(f"â€¢ í‰ê·  ë§Œì¡±ë„: {avg_satisfaction:.1f}ì ")
    print(f"â€¢ í‰ê·  ì—°ë´‰: {avg_salary/10000:.0f}ë§Œì›")


# âœ… å•é¡Œ 3-14: æœ€çµ‚å‹§å‘Šæ¡ˆä½œæˆ ğŸ’¼
print("\n--- å•é¡Œ 3-14 ---")
print("ğŸ“‹ CEO ë³´ê³ ì„œ: ì§ì› ë§Œì¡±ë„ ë° ì´ì§ ë°©ì§€ ì „ëµ")
print("=" * 60)

# TODO: í˜„í™© ìš”ì•½ ì‘ì„±í•˜ê¸°
print("\nğŸ“Š í˜„í™© ìš”ì•½")
print(f"â€¢ ì „ì²´ ì§ì› ë§Œì¡±ë„: {df_hr['ë§Œì¡±ë„'].mean():.1f}/10ì ")
print(f"â€¢ í˜„ì¬ ì´ì§ ì˜í–¥ë¥ : {df_hr['ì´ì§ì˜í–¥'].mean():.1%}")
print(f"â€¢ ì´ì§ ê³ ìœ„í—˜êµ° (ì˜ˆì¸¡í™•ë¥  70% ì´ìƒ): {len(high_risk_employees)}ëª…")

# TODO: ë¶€ì„œë³„ ìœ„í—˜ë„ ë¶„ì„í•˜ê¸°
print("\nğŸ¢ ë¶€ì„œë³„ ìœ„í—˜ë„ ë¶„ì„")
dept_risk = df_hr.groupby('ë¶€ì„œ')['ì´ì§ì˜í–¥'].mean().sort_values(ascending=False)

for dept, risk in dept_risk.items():
    if risk > 0.3:  # 30% ì´ìƒì´ë©´ ë†’ì€ ìœ„í—˜
        emoji = "ğŸ”´"
        level = "ë†’ìŒ"
    elif risk > 0.15:  # 15% ì´ìƒì´ë©´ ë³´í†µ
        emoji = "ğŸŸ¡"
        level = "ë³´í†µ"
    else:
        emoji = "ğŸŸ¢"
        level = "ë‚®ìŒ"

    print(f"{emoji} {dept}ë¶€: ì´ì§ë¥  {risk:.1%} (ìœ„í—˜ë„: {level})")

# TODO: í•µì‹¬ ë°œê²¬ì‚¬í•­ ì •ë¦¬í•˜ê¸°
print(f"\nğŸ’¡ í•µì‹¬ ë°œê²¬ì‚¬í•­")
most_important_satisfaction = coefficients.iloc[0]
most_important_turnover = odds_analysis.iloc[0]
print(f"1ï¸âƒ£ ë§Œì¡±ë„ì— ê°€ì¥ í° ì˜í–¥ì„ ë¯¸ì¹˜ëŠ” ìš”ì†ŒëŠ” '{most_important_satisfaction['ë³€ìˆ˜']}'ì…ë‹ˆë‹¤.")
print(f"2ï¸âƒ£ ì´ì§ ì˜í–¥ì— ê°€ì¥ í° ì˜í–¥ì„ ë¯¸ì¹˜ëŠ” ìš”ì†ŒëŠ” '{most_important_turnover['ë³€ìˆ˜']}'ì…ë‹ˆë‹¤.")


# TODO: ê¶Œì¥ ì•¡ì…˜ í”Œëœ ì‘ì„±í•˜ê¸°
print(f"\nğŸ¯ ê¶Œì¥ ì•¡ì…˜ í”Œëœ")
highest_risk_dept = dept_risk.index[0]  # ê°€ì¥ ìœ„í—˜í•œ ë¶€ì„œ
print(f"ğŸš¨ ì¦‰ì‹œ ì¡°ì¹˜ í•„ìš”: '{highest_risk_dept}'ë¶€ì„œì— ëŒ€í•œ ì‹¬ì¸µ ì¸í„°ë·° ë° ì›ì¸ ë¶„ì„ ì°©ìˆ˜")
print(f"ğŸ’° íˆ¬ì ìš°ì„ ìˆœìœ„: '{most_important_satisfaction['ë³€ìˆ˜']}' ê´€ë ¨ ì œë„ ê°œì„  ë° íˆ¬ì í™•ëŒ€ (ì˜ˆ: ì—°ë´‰ í˜‘ìƒ, ì§ê¸‰ ì²´ê³„ ê°œí¸)")
print(f"ğŸ‘¥ ì§‘ì¤‘ ê´€ë¦¬: ì´ì§ ê³ ìœ„í—˜êµ°ìœ¼ë¡œ ë¶„ë¥˜ëœ {len(high_risk_employees)}ëª…ì— ëŒ€í•œ ê°œë³„ ë©´ë‹´ ë° ì¼€ì–´ í”„ë¡œê·¸ë¨ ê°€ë™")
print(f"ğŸ“Š ëª¨ë‹ˆí„°ë§: ë¶„ê¸°ë³„ ì§ì› ë§Œì¡±ë„ ë° ì´ì§ ì˜í–¥ ì¡°ì‚¬ë¥¼ ì •ë¡€í™”í•˜ì—¬ ë³€í™” ì¶”ì´ ì§€ì†ì  ê´€ì°°")

print("\n", "="*20, "ë¯¸ì…˜ 3 ì¢…ë£Œ", "="*20, "\n")
```
