```python
try:
    # 'category_name'의 결측치를 특정 값으로 채웁니다.
    df['category_name'] = df['category_name'].fillna('Other/Other/Other')

    # 카테고리 분리 함수 정의
    def split_cat(category_name):
        try:
            # '/'를 기준으로 텍스트를 분리합니다.
            cats = category_name.split('/')
            # 카테고리가 3개가 아닌 경우를 대비하여 길이를 맞춥니다.
            while len(cats) < 3:
                cats.append('None')
            return cats[0], cats[1], cats[2]
        except:
            return 'Other', 'Other', 'Other'

    # apply 함수를 사용하여 카테고리를 대/중/소로 분리하고 새 열을 만듭니다.
    df['main_cat'], df['sub1_cat'], df['sub2_cat'] = \
        zip(*df['category_name'].apply(split_cat))

    print("=== 카테고리 분리 후 데이터 확인 (상위 5개) ===")
    print(df[['category_name', 'main_cat', 'sub1_cat', 'sub2_cat']].head())

    # 대분류(main_cat) 품목 분석 및 시각화
    print("\n=== 대분류 품목 개수 (상위 10개) ===")
    print(df['main_cat'].value_counts().head(10))

    plt.figure(figsize=(12, 8))
    sns.countplot(y='main_cat', data=df, order=df['main_cat'].value_counts().head(20).index)
    plt.title('상위 20개 대분류 품목')
    plt.xlabel('상품 수')
    plt.ylabel('대분류 카테고리')
    plt.show()

except NameError:
    print("NameError: 'df' is not defined. 이전 셀을 실행하여 데이터를 먼저 로드해주세요.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
try:
    # 1. 대분류 카테고리별 중앙값(median) 가격 계산 및 출력
    print("=== 대분류 카테고리별 중앙값 가격 (상위 10개) ===")
    # 가격이 0인 경우를 제외하고 계산하여 더 의미있는 통계를 얻습니다.
    median_price_by_cat = df[df['price'] > 0].groupby('main_cat')['price'].median()
    print(median_price_by_cat.sort_values(ascending=False).head(10))

    # 2. 상위 카테고리별 가격 분포 시각화 (박스 플롯)
    # 거래가 가장 많은 상위 10개 카테고리 선정
    top_10_cats = df['main_cat'].value_counts().head(10).index

    plt.figure(figsize=(14, 8))
    # 로그 변환된 가격(np.log1p)을 사용하여 가격 분포를 더 명확하게 시각화
    # 상위 10개 카테고리에 해당하는 데이터만 필터링하여 사용
    sns.boxplot(x='main_cat', y=np.log1p(df['price']), data=df[df['main_cat'].isin(top_10_cats)], order=top_10_cats)
    
    plt.title('상위 10개 대분류 카테고리별 가격 분포 (로그 변환)', fontsize=15)
    plt.xlabel('대분류 카테고리', fontsize=12)
    plt.ylabel('Log(Price + 1)', fontsize=12)
    plt.xticks(rotation=45, ha='right') # 카테고리 이름이 길 경우를 대비해 라벨을 회전
    plt.show()

except NameError:
    print("NameError: 'df' is not defined. 이전 셀들을 먼저 실행해주세요.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
try:
    # 분석할 주요 대분류 카테고리 선정 (가장 빈도가 높은 상위 3개)
    top_main_cats = df['main_cat'].value_counts().head(3).index
    print(f"분석 대상 상위 카테고리: {top_main_cats.tolist()}\n")

    # 각 대분류 카테고리별로 분석 수행
    for category in top_main_cats:
        print(f"--- 대분류: '{category}' 분석 ---")
        
        # 해당 카테고리에 속하는 데이터만 필터링
        cat_df = df[df['main_cat'] == category]
        
        # 1. 해당 대분류 내의 상위 중분류(sub1_cat) 품목
        print(f"\n[ {category} ] 내 상위 5개 중분류:")
        top_sub1_cats = cat_df['sub1_cat'].value_counts().head(5)
        print(top_sub1_cats)
        
        # 2. 해당 대분류 내의 상위 브랜드(brand_name)
        # 브랜드가 없는 경우(결측치)는 제외하고 계산
        print(f"\n[ {category} ] 내 상위 5개 브랜드:")
        top_brands = cat_df.dropna(subset=['brand_name'])['brand_name'].value_counts().head(5)
        print(top_brands)
        
        # 3. 해당 대분류 내의 상위 상품명(name)
        print(f"\n[ {category} ] 내 가장 많이 등록된 상품명 5개:")
        top_names = cat_df['name'].value_counts().head(5)
        print(top_names)
        
        print("\n" + "="*40 + "\n")

except NameError:
    print("NameError: 'df' is not defined. 이전 셀들을 먼저 실행해주세요.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
try:
    # 1. 가장 많이 등록된 품목 50개
    print("=== 1. 가장 많이 등록된 품목 (상위 50개) ===")
    most_frequent_items = df['name'].value_counts().head(50)
    print(most_frequent_items)
    print("\n" + "="*50 + "\n")

    # 2. 가장 비싼 품목 50개
    print("=== 2. 가장 비싼 품목 (상위 50개) ===")
    # 'price'를 기준으로 내림차순 정렬하여 상위 50개 선택
    most_expensive_items = df.sort_values(by='price', ascending=False).head(50)
    # 주요 정보(이름, 가격, 카테고리, 브랜드)만 출력
    print(most_expensive_items[['name', 'price', 'category_name', 'brand_name']])
    print("\n" + "="*50 + "\n")

    # 3. 가장 저렴한 품목 50개 (0원 제외)
    print("=== 3. 가장 저렴한 품목 (상위 50개, 0원 제외) ===")
    # 0원인 상품은 제외하고 가격이 가장 낮은 순으로 정렬
    cheapest_items = df[df['price'] > 0].sort_values(by='price', ascending=True).head(50)
    print(cheapest_items[['name', 'price', 'category_name', 'brand_name']])
    print("\n" + "="*50 + "\n")

    # 4. 중간 가격대 품목 50개
    print("=== 4. 중간 가격대 품목 (50개) ===")
    # 0원을 제외한 가격의 중앙값 계산
    median_price = df[df['price'] > 0]['price'].median()
    print(f"중앙값(Median Price): ${median_price:.2f}")
    
    # 중앙값과의 차이를 계산하여 새로운 열 추가
    df['price_diff'] = abs(df['price'] - median_price)
    
    # 중앙값에 가장 가까운 품목 50개 선택
    median_items = df.sort_values(by='price_diff').head(50)
    print(median_items[['name', 'price', 'category_name', 'brand_name']])
    
    # 분석 후 임시 열 삭제
    df.drop(columns=['price_diff'], inplace=True)
    print("\n" + "="*50 + "\n")

except NameError:
    print("NameError: 'df' is not defined. 이전 셀들을 먼저 실행해주세요.")
except KeyError as e:
    print(f"KeyError: {e}. 데이터에 해당 열이 없는지 확인해주세요.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
try:
    # --- 1. 상위 50개 카테고리 분석 ---
    print("=== 상위 50개 카테고리 분석 ===")
    
    # 1.1. 가장 많이 등장하는 상위 50개 카테고리 이름 추출
    top_50_categories = df['category_name'].value_counts().head(50).index.tolist()
    
    # 1.2. 해당 카테고리에 속하는 데이터만 필터링
    top_cat_df = df[df['category_name'].isin(top_50_categories)]
    
    # 1.3. 상위 50개 카테고리 내에서 가격(중앙값)이 높은 순으로 정렬하여 출력
    print("\n--- 인기(상위 50개) 카테고리의 가격 순위 (중앙값 기준) ---")
    # 가격이 0인 상품은 제외하고 중앙값을 계산하여 더 정확한 순위를 파악
    median_price_by_top_cat = top_cat_df[top_cat_df['price'] > 0].groupby('category_name')['price'].median().sort_values(ascending=False)
    print(median_price_by_top_cat)
    
    print("\n" + "="*60 + "\n")

    # --- 2. 상위 50개 브랜드 분석 ---
    print("=== 상위 50개 브랜드 분석 ===")
    
    # 2.1. 가장 많이 등장하는 상위 50개 브랜드 이름 추출 (결측치 제외)
    top_50_brands = df.dropna(subset=['brand_name'])['brand_name'].value_counts().head(50).index.tolist()
    
    # 2.2. 해당 브랜드에 속하는 데이터만 필터링
    top_brand_df = df[df['brand_name'].isin(top_50_brands)]
    
    # 2.3. 상위 50개 브랜드 내에서 가격(중앙값)이 높은 순으로 정렬하여 출력
    print("\n--- 인기(상위 50개) 브랜드의 가격 순위 (중앙값 기준) ---")
    median_price_by_top_brand = top_brand_df[top_brand_df['price'] > 0].groupby('brand_name')['price'].median().sort_values(ascending=False)
    print(median_price_by_top_brand)
    
    print("\n" + "="*60 + "\n")

except NameError:
    print("NameError: 'df' is not defined. 이전 셀들을 먼저 실행해주세요.")
except KeyError as e:
    print(f"KeyError: {e}. 데이터에 해당 열이 없는지 확인해주세요.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
try:
    # 1. 컬럼별 결측치 개수 계산
    missing_values = df.isnull().sum()
    
    # 2. 결측치가 하나 이상 있는 컬럼만 필터링
    missing_values = missing_values[missing_values > 0]
    
    # 3. 결측치 비율 계산
    missing_percent = (missing_values / len(df)) * 100
    
    # 4. 결과를 보기 좋게 데이터프레임으로 만들기
    missing_info = pd.DataFrame({
        '결측치 개수': missing_values,
        '결측치 비율 (%)': missing_percent
    })
    
    # 5. 결측치 개수가 많은 순서대로 정렬
    missing_info = missing_info.sort_values(by='결측치 개수', ascending=False)
    
    # 6. 결과 출력
    print("=== 컬럼별 결측치 정보 ===")
    if missing_info.empty:
        print("데이터에 결측치가 없습니다.")
    else:
        print(missing_info)

    # 7. 결측치 분포 시각화 (Heatmap)
    # 결측치가 있는 컬럼이 있을 경우에만 시각화 실행
    if not missing_info.empty:
        plt.figure(figsize=(12, 8))
        sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
        plt.title('데이터 결측치 분포 시각화')
        plt.show()

except NameError:
    print("NameError: 'df' is not defined. 이전 셀을 실행하여 데이터를 먼저 로드해주세요.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
# 분석할 키워드를 변수로 지정하여 쉽게 변경할 수 있도록 합니다.
keyword = 'Bundle'

try:
    # --- 1. 키워드 포함 상품 데이터 필터링 ---
    # 상품명에 특정 키워드가 포함된 데이터만 추출합니다. (대소문자 무시)
    keyword_df = df[df['name'].str.contains(keyword, case=False, na=False)].copy()
    print(f"=== '{keyword}' 키워드 포함 상품 카테고리 비교 분석 ===")
    print(f"총 {len(keyword_df)}개의 '{keyword}' 상품이 있습니다.\n")

    # --- 2. 키워드 상품의 카테고리별 가격 분포 분석 및 시각화 ---
    print(f"--- '{keyword}' 상품의 상위 카테고리별 가격 분포 ---")
    
    # 키워드 상품 중 가장 빈도가 높은 상위 10개 대분류 카테고리 선정
    top_cats_in_keyword_df = keyword_df['main_cat'].value_counts().head(10).index
    
    plt.figure(figsize=(20, 8))
    
    # 시각화 1: 키워드 상품의 카테고리별 가격 분포 (Boxplot)
    plt.subplot(1, 2, 1)
    sns.boxplot(
        x='main_cat', 
        y=np.log1p(keyword_df['price']), 
        data=keyword_df[keyword_df['main_cat'].isin(top_cats_in_keyword_df)],
        order=top_cats_in_keyword_df
    )
    plt.title(f"'{keyword}' 상품의 상위 카테고리별 가격 분포", fontsize=15)
    plt.xlabel("대분류 카테고리", fontsize=12)
    plt.ylabel("Log(Price + 1)", fontsize=12)
    plt.xticks(rotation=45, ha='right')

    # --- 3. 전체 카테고리 내 키워드 상품 비율 분석 및 시각화 ---
    print(f"\n--- 전체 상품 대비 '{keyword}' 상품 비율 (상위 카테고리) ---")

    # 전체 데이터에서 가장 빈도가 높은 상위 10개 대분류 카테고리 선정
    top_overall_cats = df['main_cat'].value_counts().head(10).index
    
    # 각 카테고리별 전체 상품 수
    category_counts = df['main_cat'].value_counts().loc[top_overall_cats]
    
    # 각 카테고리별 키워드 상품 수
    keyword_counts_in_cats = keyword_df[keyword_df['main_cat'].isin(top_overall_cats)]['main_cat'].value_counts()
    
    # 키워드 상품 비율 계산 (키워드 상품 수 / 전체 상품 수 * 100)
    # .reindex를 사용하여 두 Series의 인덱스를 맞추고, 없는 값은 0으로 채웁니다.
    keyword_proportion = (keyword_counts_in_cats.reindex(top_overall_cats, fill_value=0) / category_counts) * 100
    keyword_proportion = keyword_proportion.sort_values(ascending=False)
    
    print(keyword_proportion.round(2))

    # 시각화 2: 카테고리별 키워드 상품 비율 (Barplot)
    plt.subplot(1, 2, 2)
    sns.barplot(x=keyword_proportion.index, y=keyword_proportion.values, palette='coolwarm')
    plt.title(f"상위 카테고리 내 '{keyword}' 상품 비율 (%)", fontsize=15)
    plt.xlabel("대분류 카테고리", fontsize=12)
    plt.ylabel(f"'{keyword}' 상품 비율 (%)", fontsize=12)
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

except NameError:
    print("NameError: 'df' is not defined. 이전 셀들을 먼저 실행해주세요.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```
